```python
#Bibliotecas utilizadas
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Função para o cálculo da convolução: s[n] (sinal de entrada) e h[n] (resposta ao impulso)
#Letra (e)

def convolution(s,h): 
    K = len(s) #comprimento da matriz s[n]
    D = len(h) #comprimento da matriz h[n]
    soma = 0
    y = list(range(K+D-1))
    for n in range(0,K+D-1): #bloco condicionado pelo comprimento do canal de saída x[n]
        for k in range(0,K):
          if ((n - k) > -1) & (n - k < D):
                soma = soma + s[k]*h[n-k] #função somatório
        y[n] = round(soma,6)
        soma = 0
    return y # Retorna o valor do somatório referente à convolução

w1 = [1, 0.5, 0.5**2, 0.5**3, 0.5**4] #filtro equalizador 1
w2 = [1, 1.5, 0.7, -0.2, 0.3] #filtro equalizador 2
h = [1, -0.5] #resposta ao impulso

g1 = convolution(w1,h)
g2 = convolution(w2,h)

print("g1[n]=",g1)
print("g2[n]=",g2)
```

```python
#Letra f

s = np.random.uniform(0,1,100)
s[np.where(s < 0.25)] = 3
s[np.where(s < 0.5)] = 1
s[np.where(np.abs(s) < 0.75)] = -1
s[np.where(np.abs(s) < 1)] = -3
x = convolution(s,h)
print(x)
print(s)

plt.hist(x,color='cornflowerblue')
plt.hist(s,color='salmon')
```

```python
#Letra g

z = list(range(100))

y1 = convolution(x,w1)
y2 = convolution(x,w2)

print("\ny1:\n",y1)
print("\ny2:\n",y2)

#Plotando os gráficos

plt.figure()
plt.subplots_adjust(wspace = 0.5)
f, axs = plt.subplots(1,2,figsize=(15,15))

#Gráfico 1

plt.subplot(2,2,1)
plt.title('Análise entre s[n] e y1[n]')
plt.xlabel('n')
plt.ylabel('s[n] e y1[n]')
plt.stem(z,s[0:100], linefmt = '-.', basefmt = 'k', use_line_collection = True)
plt.stem(z,y1[0:100], linefmt =(':','r'), basefmt = 'k', markerfmt = ('o','r'), use_line_collection = True)

#Gráfico 2

plt.subplot(2,2,2)
plt.title('Análise entre s[n] e y2[n]')
plt.xlabel('n')
plt.ylabel('s[n] e y2[n]')
plt.stem(z,s[0:100], basefmt = 'k', use_line_collection = True, label = 's[n]')
plt.stem(z,y2[0:100], 'red', linefmt =(':','red'), markerfmt = ('o','red'), basefmt = 'k',use_line_collection = True,\
         label = 'y[n]')

plt.legend(loc='upper right')
```
